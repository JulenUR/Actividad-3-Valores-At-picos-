# Base de datos 2020

```python
#Se verifica la información del dataframe
data2020.info()
```

```python
#Se crea el filtro 2020
Cuant_202O = data2020.iloc[ : , [8,9,10,11,12,13,14]]
Cual_2020 = data2020.iloc[ : , [0,1,2,3,4,5,6,7,15,16,17]]
```

```python
#Se corroboran los valores nulos de las columnas con valores cuantitativos
valores_nulos = Cuant_202O.isnull().sum()
valores_nulos
```

```python
#Se reemplazan los valores nulos en dichas columnas con bfill
Cuant_202O = Cuant_202O.fillna(method = 'bfill')
```

```python
#Se corrobora que se hayan rellenado los valores nulos
valores_nulos = Cuant_202O.isnull().sum()
valores_nulos
```

```python
#Se aplica el método desviación estandar
y = Cuant_202O
Limite_Superior = y.mean() + 3 * y.std()
Limite_Inferior = y.mean() - 3 * y.std()
print("Limite Superior Permitido"), Limite_Superior
print("Limite Inferior Permitido"), Limite_Inferior
```

```python
#Se encuentran outliers del dataframe
outliers = Cuant_202O[ (y > Limite_Superior) | (y < Limite_Inferior)]
outliers
```

```python
#Se obtienen datos y los outliers se convierten en nulos en el dataframe
Cuant_202O = Cuant_202O[(y <= Limite_Superior) & (y >= Limite_Inferior)]
Cuant_202O
```

```python
#Se corroboran los valores nulos
valores_nulos = Cuant_202O.isnull().sum()
valores_nulos
```

```python
#Se reemplazan los valores nulos del dataframe con "mean"
data_clean_Cuant_2020 = Cuant_202O.copy()
data_clean_Cuant_2020 = data_clean_Cuant_2020.fillna(round(Cuant_202O.mean(), 1))
data_clean_Cuant_2020
```

```python
#Se corroboran los valores nulos
valores_nulos = data_clean_Cuant_2020.isnull().sum()
valores_nulos
```

```python
#Se corroboran los valores nulos
valores_nulos = Cual_2020.isnull().sum()
valores_nulos
```

```python
data_clean_Cual_2020 = Cual_2020.copy()
data_clean_Cual_2020 = data_clean_Cual_2020.fillna('X')
data_clean_Cual_2020
```

```python
#Se corroboran los valores nulos después de haber reemplazado
valores_nulos = data_clean_Cual_2020.isnull().sum()
valores_nulos
```

```python
#Se unen los data frames
defclean_2020 = pd.concat([data_clean_Cual_2020, data_clean_Cuant_2020], axis = 1)
defclean_2020
```

```python
defclean_2020.isnull().sum()
```

```python
#Se corroboran los valores nulos
valores_nulos = Cuant_2021.isnull().sum()
valores_nulos
```

```python
#Se encuentran los outliers en el dataframe
outliers_iqr = Cuant_2021[(y > Limite_Superior_iqr) | (y < Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Se obtienen los datos limpios del dataframe
Cuant_2021_iqr = Cuant_2021[(y <= Limite_Superior_iqr) & (y >= Limite_Inferior_iqr)]
Cuant_2021_iqr
```

```python
#Se corroboran los valores nulos
valores_nulos = Cuant_2021_iqr.isnull().sum()
valores_nulos
```

```python
#Se reemplazan los valores nulos del data frame con "mean"
data_clean_Cuant_2021=Cuant_2021.copy()
data_clean_Cuant_2021=data_clean_Cuant_2021.fillna(round(Cuant_2021.mean(), 1))
data_clean_Cuant_2021
```

```python
#Se corroboran los valores nulos
valores_nulos = data_clean_Cuant_2021.isnull().sum()
valores_nulos
```

```python
#Se corroboran los valores nulos
valores_nulos = Cual_2021.isnull().sum()
valores_nulos
```

```python
#Se corroboran los valores nulos
valores_nulos = data_clean_Cual_2021.isnull().sum()
valores_nulos
```

```python
#Se unen los dos dataframes
defclean_2021 = pd.concat([data_clean_Cual_2021, data_clean_Cuant_2021], axis = 1)
defclean_2021
```

```python
defclean_2021.isnull().sum()
```
